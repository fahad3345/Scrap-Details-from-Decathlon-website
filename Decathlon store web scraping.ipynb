```python
# Define a regular expression to extract the pin code
pincode_regex = r'\b\d{6}\b'

# Define a geolocator object to access Nominatim API
geolocator = Nominatim(user_agent="geoapiExercises")

main_list = []
#main scraping process is starting here
for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    a = soup.find_all('div', class_='content-detail')
    list1 = []
    for i in a:
        li = i.get_text(separator='\n').strip()
        list1.append(li)
    if len(list1) >= 3:
        list1[2] = list1[2].split('\n', 1)
        mob = list1[0]
        name = list1[2][0]
        address = list1[2][1].replace('\n', ',')
        timing = list1[3].replace('\n', ',')
        pincode = re.findall(pincode_regex, address)[0] if re.findall(pincode_regex, address) else pd.NA
        if pd.notna(pincode):         #we are going to find latitude and longitude based on pincode
            pincode = int(str(pincode)[:6])
            location = geolocator.geocode({"postalcode": pincode, "country": "India"})
            if location is not None:
                latitude = location.latitude
                longitude = location.longitude
            else:
                latitude = None
                longitude = None
        else:
            pincode = pd.NA
            latitude = None
            longitude = None
        main_list.append([name, address, timing, mob, pincode, latitude,longitude])
    else:
        # handle the case where there are not enough items in list1
        pass
        
data=pd.DataFrame(main_list,columns=['Name','Address','Timing','Mobile.no','Pincode','Latitude','Longitude'])
data.drop('Pincode',axis=1,inplace=True)  #we don't need it in output
data.to_csv('Retail_store.csv',index=False)   #creating a csv file
```
